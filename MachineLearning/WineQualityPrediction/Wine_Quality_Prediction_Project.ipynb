```python
#import libraries
import numpy as np
import pandas as pd

#import data visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing train test split function
from sklearn.model_selection import train_test_split
```

```python
#Import accuracy score function
from sklearn.metrics import accuracy_score
```

```python
# model selection
from sklearn.ensemble import RandomForestClassifier
```

```python
#Loading the dataset
Wine = pd.read_csv('winequality-red.csv')
Wine
```

```python
#categories the values using category plot
sns.catplot(x= 'quality' ,data=Wine, kind= 'count')
```

```python
#Splitting input features

X = Wine.drop(columns='quality', axis = 1)   #Input feature(Except the quality column all are values of x )
Y = Wine['quality']   #Output feature
```

```python
print(X)
```

```python
print(Y)
```

```python
777777777777777777777777777777777777777#Splitting the data into training data and testing data
#Training data - X_train , Y_train
#Testing data - X_test, Y_test
X_train , X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, random_state = 2)
print(X.shape, X_train.shape, Y_test.shape)
```

```python
#training the model
model = RandomForestClassifier()
```

```python
#Fit the model
model.fit(X_train, Y_train)
```

```python
#Evaluating the Trained Model
X_train_prediction = model.predict(X_train)
```

```python
#Finiding the accuracy score on training data
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print("Accuracy score of training data",training_data_accuracy)
```

```python
#Evaluating the Testing Model
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(Y_test, X_test_prediction)
#Finding the accuracy score on training data
print("Accuracy Score of Testing data",testing_data_accuracy)
```
